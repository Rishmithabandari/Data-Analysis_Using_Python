```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('insurance.csv')
df
```

```python
age_vector = df['age'].values
print('Vetor: ',age_vector)
```

```python
smoker_factor = pd.Categorical(df['smoker'])
print("Factor: ",smoker_factor)
```

```python
regions_list = df['region'].unique().tolist()
print("List: ",regions_list)
```

```python
age_bmi_matrix = df[['age', 'bmi']].values
print("Matrix: ",age_bmi_matrix)
```

```python
subset_df = df[['age', 'smoker', 'expenses']]
print("DataFrame: ",subset_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['expenses'], marker='.', color='red', label='Data points')
plt.title('Age vs. Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(df['age'], df['smoker'], marker='.', label='Data points')
plt.title('Age vs. Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoker (yes/no)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
filtered_df = df[(df['age'] >= 20) & (df['age'] <= 40)]
plt.figure(figsize=(4, 3))
plt.scatter(filtered_df['age'], filtered_df['smoker'], marker='.', label='Data points')
plt.title('Age (20-40) vs. Smoking Status')
plt.xlabel('Age')
plt.ylabel('Smoker (yes/no)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
smoker_expences = df[df['smoker']=='yes']['expenses']
non_smoker_expences = df[df['smoker']=='no']['expenses']

plt.figure(figsize=(5, 4))
plt.scatter(df['smoker'], df['expenses'], marker='.', label='Data points')
plt.title('Smoking Status vs. Expenses')
plt.xlabel('Smoker (yes/no)')
plt.ylabel('Expenses')
plt.legend()
plt.grid(True)
```

```python
smokers_gender = df[df['smoker'] == 'yes']['sex'].value_counts()
smokers_gender.plot(kind='bar', stacked=True)
plt.title('Smokers by Gender')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
```
